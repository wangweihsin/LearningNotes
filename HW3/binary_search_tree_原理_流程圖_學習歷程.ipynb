Insert -- 最終解

```python
class TreeNode(object):
    def __init__(self,x):
        self.val = x
        self.left = None
        self.right = None
        """
        :type val: int
        :type left: TreeNode or None
        :type right: TreeNode or None
        """
class Solution(object):
    def insert(self, root, val):
        if val <= root.val :
            if root.left == None :
                root.left = TreeNode(val)
                return root.left
            else:
                return self.insert(root.left,val)   
        elif val > root.val :
            if root.right == None :
                root.right = TreeNode(val)
                return root.right
            else:
                return self.insert(root.right,val)   
        """
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode(inserted node)
        """
        
root  = TreeNode(5)
Node1 = TreeNode(3)
Node2 = TreeNode(3)
Node3 = TreeNode(-5)
Node4 = TreeNode(8)
Node5 = TreeNode(7)
Node6 = TreeNode(6)
Node7 = TreeNode(10)
root.left=Node1
root.right=Node4
Node1.left=Node2
Node2.left=Node3
Node4.left=Node5
Node4.right=Node7
Node5.left=Node6

print("insert")
print(Solution().insert(root,4) ==root.left.right)
```

```python
class TreeNode(object):
    def __init__(self,x):
        self.val = x
        self.left = None
        self.right = None
        """
        :type val: int
        :type left: TreeNode or None
        :type right: TreeNode or None
        """
class Solution(object):
    def delete(self, root, target):
        parent = None
        node = root
 
        while node and node.val != target:
            parent = node
            if target < node.val:
                node = node.left
            elif target > node.val:
                node = node.right

        if node is None or node.val != target: 
            return root 

        elif node.left is None and node.right is None : 
            if target <= parent.val:
                parent.left = None
            else:
                parent.right = None
            return self.delete(root,target) 

    
        elif node.left != 0 and node.right is None: 
            if target <= parent.val:
                parent.left = node.left
            else:
                 parent.right = node.left
            return self.delete(root,target)
    
        elif node.left is None and  node.right != 0 :
            if target <= parent.val:
                parent.left = node.right
            else:
                parent.right = node.right
            return self.delete(root,target)      

        else: 
            delete_nodeparent = node
            delete_node = node.left
            while delete_node.right: 
                delete_nodeparent = delete_node            
                delete_node = delete_node.right

            node.val = delete_node.val 
            if delete_node.left == 0: 
                if delete_node.val < delete_nodeparent.val:
                    delete_nodeparent.right = None
                else:
                    delete_nodeparent.left = None
            elif delete_node.left != 0:  
                if delete_nodeparent.val > delete_node.val:
                    delete_nodeparent.right = delete_node.left
                elif delete_nodeparent.val < delete_node.val:
                    delete_nodeparent.left = delete_node.left
            return self.delete(root,target)
        
root=TreeNode(5)
root.left = TreeNode(3)
root.left.left = TreeNode(3)
root.left.left.left = TreeNode(-5)
root.right = TreeNode(8)
root.right.left = TreeNode(7)
root.right.right = TreeNode(10)
root.right.left.left = TreeNode(6)        
        
print('delete')
Solution().delete(root,3)
print(root.val==5 and root.left.val==-5 and root.left.left==None and root.left.right==None)
print(root.right.right.val==10 and root.right.left.val==7 and root.right.left.left.val==6)
print(root.right.right.right==None and root.right.right.left==None and root.right.left.right==None)
print(root.right.left.left.left==None and root.right.left.left.right==None and root.right.val==8)
```

## Modify--輔助工具

delete_1 ( ) ： 在BST用新數取代舊數(只取代一次)

Q1 : 在思考的時候發現modify和delete好像有點像，所以一開始**直接拿delete的程式加上insert的程式**，但發現會一直出錯

A1 : 後來發現，因為我們上面寫的delete是會一次把所有的重複值刪光，會變成等到全刪光以後才insert，但是這會出現錯誤的情況，因為全部刪調以後才插入，會使結構亂掉，經過反覆思考，發現應該要**刪一次就插入一次**，才可以同時更改結構(一次一次更改)，才不會出錯，所以把上面寫的delete，改寫成delete_1(一次刪一個值)後再和insert結合就好

Q2 : 在寫delete_1時，**要怎麼讓電腦知道已經刪除一個值了，該停下來了**?

A2 :　這裡再寫的時候也是卡很久，是過回傳root、root.val...都不行，後來**只要刪完一個的時候回傳一個信號**，讓電腦知道該停下來了即可

```python
class TreeNode(object):
    def __init__(self,x):
        self.val = x
        self.left = None
        self.right = None
        """
        :type val: int
        :type left: TreeNode or None
        :type right: TreeNode or None
        """
class TreeNode(object):
    def __init__(self,x):
        self.val = x
        self.left = None
        self.right = None
        """
        :type val: int
        :type left: TreeNode or None
        :type right: TreeNode or None
        """        
    def insert(self,val):
        print(self.val)
        if self.val > val:
            if self.left != None:
                self.left = TreeNode(val)
            else: 
                self.left.insert(val)
        elif self.val <= val:   
            if self.right != None:
                self.right = TreeNode(val)
            else: 
                self.right.insert(val)
    def modify(self, root, target, new_val):
        FIND_NEXT = 2
        while self.delete_1(root,target) == FIND_NEXT :
            self.insert(root,new_val)
        root = self.delete_1(root,target)
        return root
 
    def delete_1(self, root, target):    
        FIND_NEXT=2
        parent = None
        node = root        
    
        while node and node.val != target:
            parent = node
            if target < node.val:
                node = node.left
            elif target > node.val:
                node = node.right

        if node is None or node.val != target: 
            return root
       
        elif node.left is None and node.right is None: 
            if target < parent.val:
                parent.left = None
            else:
                parent.right = None
            return FIND_NEXT 

        elif node.left != 0 and node.right is None: 
            if target < parent.val:
                parent.left = node.left
            else:
                parent.right = node.left
            return FIND_NEXT

        elif node.left is None and node.right != 0: 
            if target < parent.val:
                parent.left = node.right
            else:
                parent.right = node.right
            return FIND_NEXT

        else: 
            delete_nodeparent = node
            delete_node = node.left
            while delete_node.left: 
                delete_nodeparent = delete_node            
                delete_node = delete_node.right

            node.val = delete_node.val 
            if delete_node.left == 0: 
                if delete_node.val < delete_nodeparent.val:
                    delete_nodeparent.right = None
                else:
                    delete_nodeparent.left = None
            elif delete_node.left != 0:  
                if delete_nodeparent.val > delete_node.val:
                    delete_nodeparent.right = delete_node.left
                elif delete_nodeparent.val < delete_node.val:
                    delete_nodeparent.left = delete_node.left
            return FIND_NEXT
```
