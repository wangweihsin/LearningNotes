## 新增
>insert():在BST插入一數

* 分為兩種狀況思考:

     1. 若要插入的值小於等於root的值，則要往左邊搜尋
  
     
         1.1 若左邊是空值，則建立一個節點 ( 此節點擁有TreeNode的特徵 )，並把要插入的值填入此節點
     
         1.2 若左邊非空值，則再次執行此程式，而此時root已跑至下一層，直到此節點的左邊不再有值時停止
  
      2. 若要插入的值大於root的值，則要往右邊搜尋                                 
     
         2.1 若右邊是空值，則建立一個節點 ( 此節點擁有TreeNode的特徵 )，並把要插入的值填入此節點 
     
         2.2 若右邊非空值，則再次執行此程式，而此時root已跑至下一層，直到此節點的右邊不再有值時停止
         
* 程式碼如下列所示 ( 截取自作業 )  

```python
def insert(self, root, val):
    if val <= root.val :
        if root.left == None :
            root.left = TreeNode(val)
            return root.left
        else:
            return self.insert(root.left,val)   
    elif val > root.val :
        if root.right == None :
            root.right = TreeNode(val)
            return root.right
        else:
            return self.insert(root.right,val) 
```

## 刪除
>delete ( ) : 在BST刪除此數(可能有多個)

*  分為三種狀況思考:

     1. 要刪除的值為葉子(最後一層，無左右子節點) → 直接刪除
      
      
      2. 要刪除的值只有一個節點 → 先和父節點比較
            
      3. 要刪除的值有左右子節點

         有兩種方法

          (1) 從左邊(root.left)，找最右邊的值(root.left.right.right...)，找左邊最大來取代

          (2) 從右邊(root.right)，找最左邊的值(root.right.left.left...)，找右邊最大來取代

         作業採(1)的做法


* 程式碼如下列所示 ( 截取自作業 )

```python
def delete(self, root, target):
    dad = None
    node = root

    # 先尋找target，並將找到的點存在node
    while node and node.val != target:
        dad = node
        if target < node.val:
            node = node.left
        elif target > node.val:
            node = node.right
    
    # 若沒找到target，直接回傳
    if node is None or node.val != target: 
        return root 
#################################################################
    # 若找到target，開始思考以下狀況
#################################################################
#狀況1 :無子節點
    elif node.left is None and node.right is None : # 直接刪除node(將其parent的指標設None)
        if target <= dad.val:
            dad.left = None
        else:
            dad.right = None
        return self.delete(root,target) #有可能等於target的不只有一個node，所以再delete()一次
#################################################################
#狀況2 : 只有一個子節點
    # 若有左子節點，但無右子子節點，刪除node(將其parent的指標設左子)
    elif node.left != 0 and node.right is None: 
        if target <= dad.val:
            dad.left = node.left 
        else:
            dad.right = node.left 
        return self.delete(root,target)
    # 若有右子節點，但無左子節點，刪除node(將其parent的指標設右子)
    elif node.left is None and node.right != 0 :
        if target <= dad.val:
            dad.left = node.right
        else:
            dad.right = node.right
        return self.delete(root,target)  
################################################################# 
#狀況3 : 若有左右子節點
    else: 
        delete_nodedad = node
        delete_node = node.left
        while delete_node.right: #找到左sub tree中，最右邊的node設為delete_Node，其父設為delete_NodeParent
            delete_nodedad = delete_node            
            delete_node = delete_node.right
        
        node.val = delete_node.val #將最右的node，覆蓋至要刪除的node
        if delete_node.left == 0: #若最右的node沒左子，則視delete_Node原為左子或右子，將其parent設為空
            if delete_node.val < delete_nodedad.val:
                delete_nodedad.right = None
            else:
                delete_nodedad.left = None
        elif delete_node.left != 0:  #若最右的node有左子，則直接接至其父
            if delete_nodedad.val > delete_node.val:
                delete_nodedad.right = delete_node.left
            elif delete_nodedad.val < delete_node.val:
                delete_nodedad.left = delete_node.left
        return self.delete(root,target)
```

## 查詢
>search ( ) : 在BST找尋一數

* 何謂尋訪traversal??

* 何謂Depth-First Search(DFS，深度優先搜尋)??
  
  
* 分為三種狀況思考: 

     1.  若要搜尋的值等於root 的值，則直接回傳root (**不return root.val是因為root會一直變更，值會不同**)


  
      2. 若要搜尋的值小於root的值，則要往左邊搜尋                       
  
         2.1 若左邊是空值，則直接回傳None，代表此樹沒有這個值                                         
     
         2.2 若左邊非空值，則再次執行此程式，此時root已跑至下一層，直到左邊的值等於要搜尋的值時停止
   
      3. 若要搜尋的值大於root的值，則要往右邊搜尋                      
  
         3.1 若右邊是空值，則直接回傳None，代表此樹沒有這個值                                   
   
         3.2 若右邊非空值，則再次執行此程式，此時root已跑至下一層，直到右邊的值等於要搜尋的值時停止
  
* 程式碼如下列所示 ( 截取自作業 )

```python
def search(self, root, target):
        if target == root.val :
            return root
        elif target < root.val :
            if root.left == None :
                return None
            else:
                return self.search(root.left,target)
        elif target > root.val :
            if root.right == None:
                return None
            else:
                return self.search(root.right,target)
```

## 修改
>modify ( ) : 在BST用新數取代舊數(可能有多個)

* 做法 : 

    **先刪再插入**
    
    但由於作業的規定是若有重複值要全部修改，因此要刪掉多次、插入多次
    
    不過本來寫的delete會把重複值全部**一起**刪光
    
    所以再多寫了一個delete_1( )，做到**一次只刪一個**再**只插入一個**，**重複n次**
    

* 程式碼如下列所示 ( 截取自作業 )

```python
def modify(self, root, target, new_val):
        FIND_NEXT = 2
        while self.delete_1(root,target) == FIND_NEXT : 
            self.insert(root,new_val) #再插入
        root = self.delete_1(root,target)
        return root
```

```python
def delete_1(self, root, target):    # 只刪除一次
        FIND_NEXT=2 #設FIND_NEXT為一個信號，用來判斷是否繼續做刪除
        dad = None
        node = root        
    # 尋找target,並將找到的點存在node
        while node and node.val != target:
            dad = node
            if target < node.val:
                node = node.left
            elif target > node.val:
                node = node.right

        if node is None or node.val != target: #沒找到target
            return root
       
        elif node.left is None and node.right is None: # 若node沒有孩子，刪除node(將其parent的指標設None)
            if target < dad.val:
                dad.left = None
            else:
                dad.right = None
            return FIND_NEXT #有可能等於target不只一個node，所以再delete()一次
        
         # 若有左子，但無右子，刪除node(將其parent的指標設左子)
        elif node.left != 0 and node.right is None:
            if target < dad.val:
                dad.left = node.left
            else:
                dad.right = node.left
            return FIND_NEXT
        
        # 若有右子節點，但無左子節點，刪除node(將其parent的指標設右子)
        elif node.left is None and node.right != 0:
            if target < dad.val:
                dad.left = node.right
            else:
                dad.right = node.right
            return FIND_NEXT

        else: # 若有左右子都有
            delete_nodedad = node
            delete_node = node.left
            while delete_node.left: #找到左sub tree中，最右邊的node設為delete_Node，其父設為delete_NodeParent
                delete_nodedad = delete_node            
                delete_node = delete_node.right

            node.val = delete_node.val #將最右邊的node，覆蓋至要刪除的node
            if delete_node.left == 0: # 若最右邊的node沒有右子，則視delete_Node原為左子或右子，將其parent設為空
                if delete_node.val < delete_nodedad.val:
                    delete_nodedad.right = None
                else:
                    delete_nodedad.left = None
            elif delete_node.left != 0:  #若最右邊的node有右子，則直接接至其父
                if delete_nodedad.val > delete_node.val:
                    delete_nodedad.right = delete_node.left
                elif delete_nodedad.val < delete_node.val:
                    delete_nodedad.left = delete_node.left
            return FIND_NEXT
```
